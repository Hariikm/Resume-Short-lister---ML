<h1 style="text-align: center;"><span style="color:red;">Creating the Ranking Matrix</span></h1>

**Here we are creating a new column called overall_score to calculate the assigned score on basis of diffrent conditions as mentioned below**

<h3>Overall score calculation matrix: </h3><br>

-> For the value points in python, ML, NLP and DL will be given a higher preference and the scores of that is multiplied with 3 to make the overall_score. <br>

-> For each skills in others column which are in the skill_set will be getting an extra point in the overall_score. <br>

-> For avialability 'no' a point will be deducted and for 'yes' a point will be awarded in the overall_scores. <br>

-> If the tenth mark is not given 3 points will be deducted. ( as it either shows negligence or 10th failure ) <br>

-> If the plus two mark is not given 2 points will be deducted. ( as it either shows negligence or 12th failure ) <br>

-> If the UG mark is not given 1 point will be deducted.

-> If someone got a PG mark 1 point will be awarded for the overall_score ( as it always an addition if someone got a PG degree ) <br>

-> We are not considering the streams and thier respective degree because as we all know the real talents may not be reflected from those.

```python
# Giving score for python, ML, NLP and DL.

df1['overall_score'] = (df1['python'] + df1['ML'] + df1['NLP'] + df1['DL']) * 3
```

```python
# Giving score for the relevant skills.

df1['overall_score'] += df1['others'].apply(lambda x: sum(1 for j in str(x).lower().split(',') if j.strip() in skills_list))
```

```python
# Giving score on the basis of availability.

df1['overall_score'] += df1.availability.apply(lambda x: 'yes' if x.split(',')[0].lower()=='yes' else 'no').apply(lambda x: -1 if x=='no' else 1)
```

```python
# Giving -3 for the 10th null values.

df1['overall_score'] -= df1['tenth'].isnull().astype(int) * 3
```

```python
# Giving -2 for the 12th null values.

df1['overall_score'] -= df1['plus_two'].isnull().astype(int) * 2
```

```python
# Giving -1 for the UG null values.

df1['overall_score'] -= df1['UG'].isnull().astype(int) * 1
```

```python
# Giving +1 for the PG grades.

df1['overall_score'] += df1['PG'].notnull().astype(int) * 1
```

<h1 style="text-align: center;"><span style="color:red;">Creating the ALGO</span></h1>

```python
# Looking the relevant skills in which we give the scoring for, ( we can change this as per our necessity )

skills_list
```

```python
# Defining the algorithm

def algo(df_new, top=10):
    
    # Renaming the columns
    
    df_new.rename(columns={'Python (out of 3)': 'python', 'Machine Learning (out of 3)': 'ML', 'Natural Language Processing (NLP) (out of 3)':'NLP', 'Deep Learning (out of 3)':'DL', \
                  'Other skills': 'others', 'Are you available for 3 months, starting immediately, for a full-time work from home internship? ': 'availability', \
                   'Degree':'degree', 'Stream':'stream', 'Current Year Of Graduation': 'grad_year', 'Performance_PG': 'PG', 'Performance_UG': 'UG', \
                   'Performance_12': 'plus_two', 'Performance_10': 'tenth'}, inplace=True)
    
    
    
    # Getting the unique skills from the other column
    
    unique_others= df_new.others.values

    low_other= []
    for i in unique_others:
        i=str(i)
        item= i.split(',')
        for j in item:
            j=j.strip()
            j=j.lower()
            low_other.append(j)

    low_other= set(low_other)
    
    
    # Creating overall_score column to calculate the score of a -
    # intern based on his resume considering the score matrix discussed above
    
    df_new['overall_score'] = (df_new['python'] + df_new['ML'] + df_new['DL']) * 3
    df_new['overall_score'] += df_new['others'].apply(lambda x: sum(1 for j in str(x).lower().split(',') if j.strip() in skills_list))
    df_new['overall_score'] += df_new.availability.apply(lambda x: 'yes' if x.split(',')[0].lower()=='yes' else 'no').apply(lambda x: -1 if x=='no' else 1)
    df_new['overall_score'] -= df_new['tenth'].isnull().astype(int) * 3
    df_new['overall_score'] -= df_new['plus_two'].isnull().astype(int) * 2
    df_new['overall_score'] -= df_new['UG'].isnull().astype(int) * 1
    df_new['overall_score'] += df_new['PG'].notnull().astype(int) * 1
    
        
    set_df= df_new.sort_values(by= 'overall_score', ascending= False).head(top)
     
    return set_df
```

```python
df2= pd.read_excel(r'Applications_for_Machine_Learning_internship_edited.xlsx')
```

**Here the algorithm works with any similar datasets and it gives the best candidates from the given data. By default it gives the top 10 candidates, We can change this as per our wish by specifying the top argument**

The algorithm syntax :

**algo(dataframe, top)**

<br>
Here the dataframe is the dataset we need to input and,
top is the number of top candidates we required ( default = 10 )

```python
# Top 2 candidates

algo(df2, 2)
```

```python
# Top 30 candidates

algo(df2, 30)
```
